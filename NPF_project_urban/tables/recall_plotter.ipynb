```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.7029411764705882],
    'rf_adb': [0.7306372549019607],
    'rf_lr': [0.6985294117647058],
    'rf_xgb': [0.7588235294117648],
    'gbm': [0.7708333333333334],
    'knn': [0.7693627450980391],
    'lr': [0.5161764705882352],
    'rf': [0.7426470588235293],
    'svm': [0.7421568627450981],
    'xgb': [0.7710784313725491]
}
# For unchanged, norm
data2 = {
    'adb': [0.7303225806451612],
    'rf_adb': [0.7427956989247311],
    'rf_lr': [0.7797849462365591],
    'rf_xgb': [0.8060215053763441],
    'gbm': [0.7363440860215054],
    'knn': [0.7987096774193547],
    'lr': [0.5118279569892473],
    'rf': [0.8124731182795699],
    'svm': [0.8064516129032258],
    'xgb': [0.7823655913978496]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Recalls of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Recall (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['Urban', 'Arctic'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('recalls.png', dpi=300)
```
