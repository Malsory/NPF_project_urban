```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'
# For changed, smote
data1 = {
    'adb': [0.7194813477230357],
    'rf_adb': [0.7337591925904621],
    'rf_lr': [0.7103623301172322],
    'rf_xgb': [0.7633939864744819],
    'gbm': [0.7702155054554436],
    'knn': [0.7843014661014391],
    'lr': [0.5251650069991031],
    'rf': [0.7541578823080373],
    'svm': [0.7445283139632984],
    'xgb': [0.7702790283904835]
}
# For unchanged, norm
data2 = {
    'adb': [0.7289490019562008],
    'rf_adb': [0.7408179425603996],
    'rf_lr': [0.7837315387226466],
    'rf_xgb': [0.8096752425430201],
    'gbm': [0.7345562890167544],
    'knn': [0.7997157931083333],
    'lr': [0.49472213909242957],
    'rf': [0.8146191873726956],
    'svm': [0.8075436876813491],
    'xgb': [0.7807894339863746]
}

df1 = pd.DataFrame.from_dict(data1, orient='index', columns=['Value']) * 100
df2 = pd.DataFrame.from_dict(data2, orient='index', columns=['Value']) * 100
# Merge the two DataFrames on common index (names)
merged_df = pd.merge(df1, df2, left_index=True, right_index=True)
merged_df['models'] = merged_df.index

# Plotting side-by-side bar plots
plt.figure()
merged_df.plot(kind='bar', width=0.9)
plt.title('Precisions of models', fontsize=40)
plt.xlabel('Models', fontsize=30)
plt.ylabel('Precision (%)', fontsize=30)
plt.xticks(rotation=45)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylim(0, 100)
plt.legend(['Urban', 'Arctic'], loc='upper right', fontsize=30)

fig = plt.gcf()
fig.set_size_inches(25, 15)

plt.savefig('precisions.png', dpi=300)
```
